```python
import numpy as np
import pandas as pd
import os
```

```python
# paths
base_image_dir = os.path.join('', 'data')
train_image_dir = os.path.join(base_image_dir, 'train')

# read the data
retina_df = pd.read_csv(os.path.join(base_image_dir, 'trainLabels.csv'))
```

```python
# Clean labels
retina_df['PatientId'] = retina_df['image'].map(lambda x: x.split('_')[0])

# Does the patient from trainLabels exists in our train.zip001
retina_df['path'] = retina_df['image'].map(lambda x: os.path.join(train_image_dir, f'{x}.jpeg'))
retina_df['exists'] = retina_df['path'].map(os.path.exists)

# Left or right?
retina_df['eye'] = retina_df['image'].map(lambda x: 1 if x.split('_')[-1]=='left' else 0)
```

```python
# How many patients from one .zip are in labels?
print(retina_df['exists'].sum(), 'images found of', retina_df.shape[0], 'total')

# Take only them
retina_df = retina_df.loc[retina_df['exists']]
```

```python
# categorical one hot encoding
one_hot = pd.get_dummies(retina_df['level'])
retina_df = retina_df.drop('level', axis = 1)
retina_df = retina_df.join(one_hot)
retina_df.columns = ['image', 'PatientId', 'path', 'exists', 'is_eye_left', 'is_level_0', 'is_level_1', 'is_level_2', 'is_level_3', 'is_level_4']
retina_df.head()
```
